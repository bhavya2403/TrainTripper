```python
import pandas as pd
import json


data_list = []





file_path = 'rough.json'  


try:
    with open(file_path, "r") as file:
        json_data = json.load(file)
        
except FileNotFoundError:
    print(f"File '{file_path}' not found.")
except json.JSONDecodeError as e:
    print(f"Error decoding JSON in the file: {e}")
except Exception as e:
    print(f"An error occurred: {e}")



for data in json_data:
    train_number = data["trainNumber"]
    train_name = data["trainName"]
    distance = data["distance"]
    duration = data["duration"]

    duration_hours = duration


    hours, minutes = duration_hours.split(":")
    hours = int(hours)
    minutes = int(minutes)


    total_minutes = hours * 60 + minutes



    
    class_prices = [{"class": item["class"], "totalFare": item["totalFare"] , "basefare" : item["baseFare"] ,"reservationCharge" : item["reservationCharge"] , "superfastCharge" : item["superfastCharge"] , "dynamicFare" : item["dynamicFare"] , "cateringCharge" : item["cateringCharge"] , "fuelAmount" : item["fuelAmount"] , "totalConcession" : item["totalConcession"] , "tatkalFare" : item["tatkalFare"] , "serviceTax" : item["serviceTax"] , "otherCharge" : item["otherCharge"]} for item in data["pricesAndAvail"]]

    df = pd.DataFrame(class_prices)
    df["total_minutes"] = total_minutes
    df["distance"] = distance
    df["train number"] = train_number
    df["train name"] = train_name
    df = df[['train number', 'train name', 'class' , 'basefare' , 'reservationCharge' , 'superfastCharge' , 'dynamicFare' , 'cateringCharge' , 'fuelAmount' , 'totalConcession' , 'tatkalFare' , 'serviceTax' , 'otherCharge'  , 'distance' , 'total_minutes', 'totalFare']]

    data_list.append(df)


combined_df = pd.concat(data_list, ignore_index=True) # ignore_index=True to reset the index of the resultant DataFrame



```

```python
dummies = pd.get_dummies(combined_df['class'])
dummies
```

```python
merged = pd.concat([combined_df, dummies], axis='columns')
merged
```

```python
final = merged.drop(['class'], axis='columns')
final
```

```python
final.to_csv("train_prices_combined.csv", index=False)
```
